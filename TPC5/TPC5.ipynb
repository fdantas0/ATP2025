```python
def criar_turma():
    return []

def adicionar_aluno(turma):
    nome = input("Nome do aluno: ")
    aluno_id = int(input("ID do aluno: "))
    nota_tpc = float(input("Nota dos TPCs: "))
    nota_proj = float(input("Nota do Projeto: "))
    nota_teste = float(input("Nota do Teste: "))

    aluno = (nome, aluno_id, [nota_tpc, nota_proj, nota_teste])
    turma.append(aluno)

    print("Aluno adicionado com sucesso!\n")
    return turma

def mostrar_turma(turma):
    if not turma:
        print("A turma ainda não tem alunos.\n")
        return
    print("\n--- Lista de alunos ---")
    for nome, aluno_id, notas in turma:
        print(f"Aluno: {nome} | ID: {aluno_id} | TPC: {notas[0]} | Projeto: {notas[1]} | Teste: {notas[2]}")
    print("-----------------------\n")

def procurar_aluno(turma, aluno_id):
    for nome, id_atual, notas in turma:
        if id_atual == aluno_id:
            print(f"Aluno encontrado: {nome} (ID: {id_atual}) -> Notas: {notas}")
            return
    print("❌ Aluno não encontrado.\n")

def guardar_turma(turma, ficheiro):
    with open(ficheiro, "w") as f:
        for nome, aluno_id, notas in turma:
            linha = f"{nome};{aluno_id};{notas[0]}::{notas[1]}::{notas[2]}\n"
            f.write(linha)
    print(f"Turma guardada no ficheiro '{ficheiro}'.\n")

def carregar_turma(ficheiro):
    turma = []
    try:
        with open(ficheiro, "r") as f:
            for linha in f:
                if linha.strip():
                    nome, aluno_id, notas_txt = linha.strip().split(";")
                    notas = [float(n) for n in notas_txt.split("::")]
                    turma.append((nome, int(aluno_id), notas))
        print("Turma carregada com sucesso!\n")
    except FileNotFoundError:
        print("Ficheiro não encontrado.\n")
    return turma

def menu():
    print("""
========== MENU ==========
1 - Criar nova turma
2 - Inserir aluno
3 - Listar turma
4 - Consultar aluno por ID
5 - Guardar turma em ficheiro
6 - Carregar turma de ficheiro
0 - Sair
==========================
""")

turma = []
opcao = -1

while opcao != 0:
    menu()
    try:
        opcao = int(input("Escolha uma opção: "))
    except ValueError:
        print("Por favor, introduza um número válido.\n")
        continue

    if opcao == 0:
        print("Programa terminado.")
    elif opcao == 1:
        turma = criar_turma()
        print("Nova turma criada!\n")
    elif opcao == 2:
        turma = adicionar_aluno(turma)
    elif opcao == 3:
        mostrar_turma(turma)
    elif opcao == 4:
        aluno_id = int(input("ID do aluno a procurar: "))
        procurar_aluno(turma, aluno_id)
    elif opcao == 5:
        guardar_turma(turma, "turma.txt")
    elif opcao == 6:
        turma = carregar_turma("turma.txt")
    else:
        print("Opção inválida. Tente novamente.\n")
```
